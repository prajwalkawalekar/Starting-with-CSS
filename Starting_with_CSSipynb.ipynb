```python
""" 1. Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the project where required. """

""" HTML file (index.html) """
<!DOCTYPE html>
<head>
    <title>CSS Example</title>
    <link rel="stylesheet" href="style.css">

    <!-- Internal CSS -->
    <style>
        /* Internal CSS */
        .internal-css {
            background-color: lightblue;
            padding: 20px;
        }
    </style>

</head>
<body>
    <!-- Inline CSS -->
    <div style="background-color: yellow; padding: 20px;">Inline CSS</div>


    <div class="internal-css">Internal CSS</div>

    <!-- External CSS -->
    <div class="external-css">External CSS</div>

    <!-- Multiple CSS Selectors -->
    <div class="multiple-css-selectors">Multiple CSS Selectors</div>

</body>
</html>

""" CSS file (style.css)"""
/* External CSS */
.external-css {
    background-color: lightgreen;
    padding: 20px;
}

/* Multiple CSS Selectors */
.multiple-css-selectors {
    background-color: orange;
    padding: 20px;
}

/* Targeting the same element with multiple selectors */
div {
    font-size: 24px;
    font-weight: bold;
}
```
